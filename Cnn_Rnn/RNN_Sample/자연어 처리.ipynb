```python
from konlpy.tag import Komoran
k = Komoran()
print(komoran.morphs('딥러닝이 쉽냐? 난 어렵던데?'))
```

```python
#텍스트에서 품사를 태킹하여 반환
print(k.pos('소파 위에 있는 것이 고양이인가요? 강아지인가요?'))
```

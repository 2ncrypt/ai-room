
크롤링 예제 사이트 코드

```python
import requests
from bs4 import  BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html')
soup = BeautifulSoup(res.content,'html.parser')

#예제1 태그 선택
items = soup.select('li')
for item in items:
    print(item.get_text())
```

```python
#예제2-1 하위태그 선택
itmes2 = soup.select('html body h1')
                    #정확하게 경로 지정을 하려면 html>head>body>h1 이런식으로 해도됨.
for item in itmes2:
    print(item.get_text())
```

```python
#예제 2-2 하위태그 선택 ('>'는 바로 아래 태그인 경우를 검색)
items2 = soup.select('ul>li')
for item in items2:
    print(item.get_text())
```

```python
#예제 2-3 '.'클래스 이름으로 검색 가능
items3 = soup.select('.course') 
                    #class course를 모두 검색해서 가져옴
for item in items3:
    print(item.get_text())
```

```python
#예제 2-4 id이름으로 검색
#html css에서 class는 '.' id는 '#' 으로 표시
items4 = soup.select('#start')
for _ in items4 :
    print(_.get_text())
```

```python
#예제 2-5 (태그.클래스이름.클래스이름2)
items5 = soup.select('li.course.paid')
for item in items5:
    print(item.getText())
```

```python
#예제 2-6(복합 예제)
items6 = soup.select('ul#hobby_course_list li.course')
for item in items6:
    print(item.getText())
```

```python
#예제 2-7 (selcet_one)
item7 = soup.select_one('ul#dev_course_list > li.course.paid')
print(item.getText())
```

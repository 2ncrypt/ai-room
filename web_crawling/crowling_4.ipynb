```python
import requests
from bs4 import  BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html')
soup = BeautifulSoup(res.content,'html.parser')

#1  
items = soup.select('li')
for item in items:
    print(item.get_text())
```

```python
#2-1  
itmes2 = soup.select('html body h1')
                    #    html>head>body>h1  .
for item in itmes2:
    print(item.get_text())
```

```python
# 2-2   ('>'     )
items2 = soup.select('ul>li')
for item in items2:
    print(item.get_text())
```

```python
# 2-3 '.'   
items3 = soup.select('.course') 
                    #class course   
for item in items3:
    print(item.get_text())
```

```python
# 2-4 id 
#html css class '.' id '#'  
items4 = soup.select('#start')
for _ in items4 :
    print(_.get_text())
```

```python
# 2-5 (..2)
items5 = soup.select('li.course.paid')
for item in items5:
    print(item.getText())
```

```python
# 2-6( )
items6 = soup.select('ul#hobby_course_list li.course')
for item in items6:
    print(item.getText())
```

```python
# 2-7 (selcet_one)
item7 = soup.select_one('ul#dev_course_list > li.course.paid')
print(item.getText())
```

```python
import requests,openpyxl
from bs4 import BeautifulSoup

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['Rank','Name','price'])
excel_sheet.column_dimensions['B'].width = 20
excel_sheet.column_dimensions['C'].width = 70
excel_sheet.column_dimensions['D'].width = 50
cap = []
pay = []
res = requests.get('https://search.musinsa.com/category/007001?device=&d_cat_cd=007001&brand=&rate=&page_kind=search&list_kind=small&sort=sale_high&sub_sort=1m&page=1&display_cnt=90&sale_goods=&ex_soldout=&color=&price1=&price2=&exclusive_yn=&size=&tags=&sale_campaign_yn=&timesale_yn=&q=')
soup = BeautifulSoup(res.content,'html.parser')
items = soup.select('#searchList > li > div.li_inner > div.article_info > p.list_info > a')
payp = soup.select('#searchList > li > div.li_inner > div.article_info > p.price')
for index,item in enumerate(items):
    cap.append(item.text)
    cap[index] = ''.join(cap[index].split())
    
for index, payp in enumerate(payp):
    pay.append(payp.text)
    pay[index] = ''.join(pay[index].split())
    print(index+1,'위 ' ,cap[index],pay[index])

excel_file.save('test.xlsx')
excel_file.close()
#searchList > li:nth-child(2) > div.li_inner > div.article_info > p.list_info > a
```

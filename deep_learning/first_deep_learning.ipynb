```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

#jupyter notebook kerner die 해결법
# import os
# os.environ[‘KMP_DUPLICATE_LIB_OK’]=‘True’
```

```python
np.random.seed(3)
tf.random.set_seed(3)

Data_set = np.loadtxt("ThoraricSurgery.csv",delimiter=",")
```

```python
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,Y ,epochs=100, batch_size=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt

import numpy as np
import tensorflow as tf
```

```python
X = data = np.linspace(1,2,200) #200행 1열
# X.shape # 만들어진 값 확인
Y = (X * 4) + (np.random.randn(200) * 0.3)

# Sequential #이걸 실행시키면 한 줄의 신경망이 만들어짐
model = Sequential()
model.add(Dense(1, input_dim=1, activation='linear')) #linear 는 선 
#modle.add()함수를 사용하면 필요한 층을 추가 할 수 있음.
model.compile(optimizer='sgd', loss='mse', metrics=['mse']) #mse는 오차인데 제곱오차
model.fit(X , Y, batch_size=1, epochs=30)
```
